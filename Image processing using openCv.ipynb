```python
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
gray=cv2.imread("thor.jpg")         # pil = Image.open("thor.jpg")
```

```python
plt.imshow(gray)
```

```python
rgb=cv2.cvtColor(gray,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(rgb)
```

```python
classifier=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces=classifier.detectMultiScale(gray)
```

```python
faces
```

```python
face=faces[0]
```

```python
x,y,w,h = face
```

```python
face
```

```python
fac=rgb[y:y+h,x:x+w]
```

```python
plt.imshow(fac)
```

```python
out=cv2.rectangle(rgb,(x,y),(x+w,y+h),(0,0,225),4)
```

```python
plt.imshow(out)
```
