# DROPPING NULL VALUES

```python
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='all',axis=1)
```

# DROPPING DUBLICATES

```python
df.drop_duplicates(subset=['toy'])
```

```python
df.drop_duplicates(subset=['toy'],keep='last')
```

```python
df.drop_duplicates(subset=['toy'],keep='first')
```
